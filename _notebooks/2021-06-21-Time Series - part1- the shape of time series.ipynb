# Time Series

   Time Series is a type of data or a problem in machine learning. In Time Series, data points are generated by the time and follow some patterns such as trending or period. Time series is very popular problem in reality, especially in trading. The price of stock or number of product sales are time series examples. 

![](time_series/stationary-1.png)

Basically, Time series is a collection of data points. Each data point has 2 attributes: value and time. The data points are sorted by their time in order to put into collection. time is something we know it but value is uncertainty and we did not know it beforehand. Then, a data point depends on its period (time)
and its trending (how big of its value). 

When the time factor does not effect to value of data points, we call this time series is stationary. Otherwise, time series is trending or seasonal. In
